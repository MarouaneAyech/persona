```python
import pandas as pd
```

```python
df=pd.read_excel('2023_2024.xlsx')
df.head()
```

```python
df.drop(['Num.','PHOTO','NOM','PRENOM','GROUPE','CIN/PAS','E-MAIL','TEL MOBILE'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['ville']=df['ADRESSE'].apply(lambda adresse : adresse.split()[-1])
df.drop('ADRESSE', axis=1, inplace=True)
df.head()
```

```python
df.drop('NIVEAU', axis=1, inplace=True)
df.head()
```

```python
from datetime import datetime

df['DATE INS.'] = pd.to_datetime(df['DATE INS.'], yearfirst=True, format='%Y-%m%d')
df['DATE INS.'].dtype
```

```python
df['Annee']=df['DATE INS.'].apply(lambda date : date.year)
df.drop('DATE INS.', axis=1, inplace=True)
df.head()
```

```python
df.set_index('CODE', inplace=True)
df.head()
```
